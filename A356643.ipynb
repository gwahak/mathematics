## A356643 : a(n) is the number of order-n magic triangles composed of the numbers from 1 to n(n+1)/2 in which the sum of the k-th row and the (n-k)-th row is same for all k and all three directions, counted up to rotations and reflections.

```python
import itertools
import numpy as np
```

```python
i=0
sum2=11
sum4=22
for A in itertools.permutations(range(1,11), 10):
    if A[1]+A[2]==sum2:
        if A[3]+A[7]==sum2:
            if A[5]+A[8]==sum2:
                if A[0]+A[1]+A[3]+A[6]==sum4:
                    if A[6]+A[7]+A[8]+A[9]==sum4:
                        if A[0]+A[2]+A[5]+A[9]==sum4:
                            i+=1
print(i)
```
