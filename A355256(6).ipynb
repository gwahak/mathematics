```python
from itertools import permutations

def find_valid_2x2_subsquares(n=36, target_sum=74):
    count = 0
    valid_tuples = []

    for A, B, C in permutations(range(1, n + 1), 3):
        if not (A < B and A < C):
            continue  # A must be the smallest of the three
        if not (B < C):
            continue  # A must be the smallest of the three
        D = target_sum - (A + B + C)

        if not (1 <= D <= n):
            continue
        if D in (A, B, C):
            continue  # All four values must be distinct
        if D<A:
            continue
        count += 1
        valid_tuples.append((A, B, C, D))
        print(A,B,C,D)
    return count, valid_tuples

count, squares = find_valid_2x2_subsquares()
print(f"✅ Number of valid 2×2 subsquares: {count}")
print("🧾 First 5 valid (A, B, C, D):", squares[:5])
```
