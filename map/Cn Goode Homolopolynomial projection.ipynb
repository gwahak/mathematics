# Cn Smooth Goode Homolopolynomial

*The classical Goode homolosine projection has a visible kink at latitude 40.736662°, because the sinusoidal and Mollweide regions are joined only with C0 continuity.

By inserting a Strebe's equal-area polynomial projection transition region between the two projections, a Cn-smooth Goode Homolopolynomial variant can be constructed.*

```python
import numpy as np
import matplotlib.pyplot as plt

# Latitude grid, avoiding exact poles for numerical stability
lat_deg = np.linspace(-89.999, 89.999, 3000)
phi = np.deg2rad(lat_deg)

# Sinusoidal: y = phi
p_sinusoidal = np.ones_like(phi)

# Mollweide:
# 2θ + sin(2θ) = π sin φ
# y = sqrt(2) sin θ
# dy/dφ = sqrt(2) cosθ dθ/dφ
# dθ/dφ = π cosφ / (2 + 2 cos 2θ)
theta = phi / 2
for _ in range(60):
    f = 2 * theta + np.sin(2 * theta) - np.pi * np.sin(phi)
    fp = 2 + 2 * np.cos(2 * theta)
    theta -= f / fp

p_mollweide = np.sqrt(2) * np.cos(theta) * (np.pi * np.cos(phi) / (2 + 2 * np.cos(2 * theta)))

# Goode homolosine junction latitude:
# cos φ0 = (2 sqrt(2) / π) cos θ0
# numerically find where Mollweide dy/dφ = 1 in positive latitude
idx_pos = np.where(lat_deg >= 0)[0]
junction_idx = idx_pos[np.argmin(np.abs(p_mollweide[idx_pos] - 1))]
junction_lat = lat_deg[junction_idx]

# ±90° 점 추가
lat_deg_plot = np.concatenate(([-90.0], lat_deg, [90.0]))

# Sinusoidal은 dy/dφ = 1 이므로 극점에서도 1
p_sinusoidal_plot = np.concatenate(([1.0], p_sinusoidal, [1.0]))

# Mollweide는 극점에서 0
pole_value = 0
p_mollweide_plot = np.concatenate(([0], p_mollweide, [0]))

plt.figure(figsize=(8, 5))
plt.plot(lat_deg_plot, p_sinusoidal_plot, label="Sinusoidal: dy/dφ = 1")
plt.plot(lat_deg_plot, p_mollweide_plot, label="Mollweide: actual dy/dφ")

plt.axvline(junction_lat, linestyle="--",
            label=f"junction ≈ {junction_lat:.4f}°")
plt.axvline(-junction_lat, linestyle="--")

plt.xlabel("Latitude (degrees)")
plt.ylabel("Parallel spacing function dy/dφ")
plt.title("Parallel spacing function: Sinusoidal vs. Mollweide")
plt.grid(True)
plt.legend()
plt.tight_layout()

path = "/mnt/data/corrected_parallel_spacing_sinusoidal_mollweide.png"
plt.savefig(path, dpi=200)
plt.show()

junction_lat, path

```

 # C¹n-smooth Goode Homolopolynomial construction

To remove the discontinuity of the Goode Homolosine projection, we employ Strebe's equal-area polynomial transition between the Sinusoidal and Mollweide projections. Depending on the construction, the polynomial is chosen to control either the parallel-spacing function itself or its derivative.

Because these are pseudocylindrical projections, equality of the \(n\)-th derivatives of the meridians at a junction point is equivalent to equality of the \(n\)-th derivatives of the parallel-spacing function at that point. Consequently, if the two projections are matched up to order \(n\), their parallel-spacing functions have identical Taylor coefficients through degree \(n\), and may differ only from degree \(n+1\) onward.

Here, the **parallel-spacing density** is defined as

$
p(\phi)=\frac{dy}{d\phi},
$

where \(y\) is the projected vertical coordinate and \(\phi\) is latitude.

For the Sinusoidal projection,

$
y=\phi,
$

and therefore

$
p(\phi)=\frac{dy}{d\phi}=1.
$

Hence the parallel-spacing density is constant, and all of its derivatives vanish:

$
p^{(k)}(\phi)=0
\qquad (k\ge 1).
$

As a result, imposing \(C^n\) continuity at the lower junction latitude requires the transition polynomial to have the same value and the same first through \(n\)-th derivatives as the constant function \(p(\phi)=1\). Equivalently, the transition polynomial and the Sinusoidal parallel-spacing density differ only in terms of degree \(n+1\) and higher in their local Taylor expansions.



 # C¹-smooth Goode Homolopolynomial construction

$
p(\phi)=1-a t^2,\qquad t=\phi-\phi_1.
$

Although the transition function contains two parameters, \(a\) and \(\phi_1\), the lower junction conditions at \(t=0\) are satisfied identically:

$
p(\phi_1)=1,\qquad p'(\phi_1)=0.
$

Thus the Sinusoidal–polynomial matching requires no additional constraints.

At the upper junction latitude \(\phi_2\), the polynomial must match the Mollweide spacing function and its first derivative:

$
p(\phi_2)=p_M(\phi_2),
\qquad
p'(\phi_2)=p_M'(\phi_2).
$

These two conditions determine \(a\) and \(\phi_1\) uniquely for a given choice of \(\phi_2\). Therefore, the admissible pairs $\((\phi_1,\phi_2)\)$ lie on a one-dimensional curve in parameter space. Consequently, the quadratic transition family has only one effective degree of freedom.

$p(\phi)=1-a t^2, \phi-\phi_1=t$



The quadratic transition function possesses one degree of freedom. By varying the upper transition latitude \(\phi_2\), one can generate a family of \(C^1\)-smooth connections between the Sinusoidal and Mollweide projections.

Nevertheless, \(C^1\) continuity alone does not guarantee satisfactory geometric properties. The transition must also preserve the desired concavity of the parallel-length function. Consequently, the admissible range of transition latitudes is more restricted than the range allowed by the continuity conditions alone.

```python
import numpy as np
from scipy.optimize import brentq, minimize_scalar
import matplotlib.pyplot as plt

# -----------------------------
# Mollweide parallel spacing p(phi)
# -----------------------------
def theta_from_phi(phi):
    phi = np.asarray(phi, dtype=float)
    theta = phi / 2
    for _ in range(80):
        f = 2*theta + np.sin(2*theta) - np.pi*np.sin(phi)
        fp = 2 + 2*np.cos(2*theta)
        theta = np.where(np.abs(fp) > 1e-13, theta - f/fp, theta)
    return theta

def p_moll(phi):
    th = theta_from_phi(phi)
    return np.pi*np.cos(phi)/(2*np.sqrt(2)*np.cos(th))

def dp_moll(phi):
    h = 1e-6
    return (p_moll(phi+h) - p_moll(phi-h))/(2*h)

# -----------------------------
# Quadratic C1 family:
# p(t)=1-a t^2, t=phi-phi1
# At phi2, P=pM(phi2), Q=pM'(phi2)
# T=-2(1-P)/Q, phi1=phi2-T, a=(1-P)/T^2
# -----------------------------
def quadratic_params_from_phi2(phi2):
    P = float(p_moll(phi2))
    Q = float(dp_moll(phi2))
    T = -2*(1-P)/Q
    phi1 = phi2 - T
    a = (1-P)/T**2
    return phi1, phi2, T, a, P, Q

def A2_max_for_phi2(phi2):
    phi1, phi2, T, a, P, Q = quadratic_params_from_phi2(phi2)
    if not (0 < phi1 < phi2 < np.pi/2 and a > 0 and T > 0):
        return np.nan, None
    
    def A(x):
        t = x - phi1
        return np.cos(x)/(1 - a*t*t)
    
    def A2(x):
        h = 1e-4
        return (A(x+h) - 2*A(x) + A(x-h))/(h*h)
    
    res = minimize_scalar(lambda z: -A2(z), bounds=(phi1 + 1e-7, phi2 - 1e-7), method="bounded",
                          options={"xatol": 1e-12})
    return -res.fun, (phi1, phi2, T, a, res.x)

# Scan phi2 and find where max A'' crosses 0
deg_grid = np.linspace(40.74, 89.0, 601)
vals = []
params = []
for deg in deg_grid:
    phi2 = np.deg2rad(deg)
    try:
        m, par = A2_max_for_phi2(phi2)
    except Exception:
        m, par = np.nan, None
    vals.append(m); params.append(par)
vals = np.array(vals)

# Find crossing from positive to non-positive
bracket = None
for i in range(len(deg_grid)-1):
    if np.isfinite(vals[i]) and np.isfinite(vals[i+1]):
        if vals[i] > 0 and vals[i+1] <= 0:
            bracket = (np.deg2rad(deg_grid[i]), np.deg2rad(deg_grid[i+1]))
            break

print("bracket deg:", None if bracket is None else (np.rad2deg(bracket[0]), np.rad2deg(bracket[1])))

root_phi2 = brentq(lambda x: A2_max_for_phi2(x)[0], bracket[0], bracket[1], xtol=1e-11)
maxA2, par = A2_max_for_phi2(root_phi2)
phi1, phi2, T, a, touch = par

print("critical upper phi2 deg =", np.rad2deg(phi2))
print("critical lower phi1 deg =", np.rad2deg(phi1))
print("transition width T deg =", np.rad2deg(T))
print("a =", a)
print("touch latitude deg =", np.rad2deg(touch))
print("max A'' =", maxA2)
print("p(t)=1-a t^2; t in radians")

# Check sample A'' min/max across interval
xs = np.linspace(phi1, phi2, 3000)
def A_quad(x):
    t = x - phi1
    return np.cos(x)/(1-a*t*t)
A = A_quad(xs)
A2_arr = np.gradient(np.gradient(A, xs), xs)
print("sample A'' min/max:", A2_arr.min(), A2_arr.max())

# Plot p and A''
lat_deg = np.linspace(np.rad2deg(phi1)-3, np.rad2deg(phi2)+3, 2000)
ph = np.deg2rad(lat_deg)
p_quad = np.full_like(ph, np.nan, dtype=float)
mask = (ph >= phi1) & (ph <= phi2)
p_quad[mask] = 1 - a*(ph[mask]-phi1)**2

plt.figure(figsize=(9,5.4))
plt.plot(lat_deg, p_moll(ph), label="Mollweide p(φ)")
plt.plot(lat_deg, np.ones_like(ph), label="Sinusoidal p=1")
plt.plot(lat_deg, p_quad, linewidth=2.6, label=r"Critical quadratic transition $1-at^2$")
plt.axvline(np.rad2deg(phi1), linestyle="--", label=f"φ1≈{np.rad2deg(phi1):.4f}°")
plt.axvline(np.rad2deg(phi2), linestyle="--", label=f"φ2≈{np.rad2deg(phi2):.4f}°")
plt.xlabel("Latitude (degrees)")
plt.ylabel("Parallel spacing p(φ)")
plt.title("First concavity-preserving C¹ quadratic transition")
plt.grid(True)
plt.legend()
plt.tight_layout()
p_path = "/mnt/data/critical_quadratic_c1_spacing.png"
plt.savefig(p_path, dpi=220)
plt.show()

plt.figure(figsize=(9,5.2))
plt.plot(np.rad2deg(xs), A2_arr, linewidth=2)
plt.axhline(0, linewidth=1)
plt.axvline(np.rad2deg(touch), linestyle=":", label=f"touch≈{np.rad2deg(touch):.4f}°")
plt.xlabel("Latitude (degrees)")
plt.ylabel("A''(φ), numerical")
plt.title(r"Concavity boundary for $A(\phi)=\cos\phi/p(\phi)$")
plt.grid(True)
plt.legend()
plt.tight_layout()
a2_path = "/mnt/data/critical_quadratic_c1_A_second_derivative.png"
plt.savefig(a2_path, dpi=220)
plt.show()

p_path, a2_path
```

```python
import numpy as np
from scipy.optimize import brentq, minimize_scalar
import matplotlib.pyplot as plt

# -----------------------------
# Mollweide parallel spacing p(phi)
# -----------------------------
def theta_from_phi(phi):
    phi = np.asarray(phi, dtype=float)
    theta = phi / 2
    for _ in range(80):
        f = 2*theta + np.sin(2*theta) - np.pi*np.sin(phi)
        fp = 2 + 2*np.cos(2*theta)
        theta = np.where(np.abs(fp) > 1e-13, theta - f/fp, theta)
    return theta

def p_moll(phi):
    th = theta_from_phi(phi)
    return np.pi*np.cos(phi)/(2*np.sqrt(2)*np.cos(th))

def dp_moll(phi):
    h = 1e-6
    return (p_moll(phi+h) - p_moll(phi-h))/(2*h)

# -----------------------------
# Quadratic C1 family:
# p(t)=1-a t^2, t=phi-phi1
# At phi2, P=pM(phi2), Q=pM'(phi2)
# T=-2(1-P)/Q, phi1=phi2-T, a=(1-P)/T^2
# -----------------------------
def quadratic_params_from_phi2(phi2):
    P = float(p_moll(phi2))
    Q = float(dp_moll(phi2))
    T = -2*(1-P)/Q
    phi1 = phi2 - T
    a = (1-P)/T**2
    return phi1, phi2, T, a, P, Q

def A2_max_for_phi2(phi2):
    phi1, phi2, T, a, P, Q = quadratic_params_from_phi2(phi2)
    if not (0 < phi1 < phi2 < np.pi/2 and a > 0 and T > 0):
        return np.nan, None
    
    def A(x):
        t = x - phi1
        return np.cos(x)/(1 - a*t*t)
    
    def A2(x):
        h = 1e-4
        return (A(x+h) - 2*A(x) + A(x-h))/(h*h)
    
    res = minimize_scalar(lambda z: -A2(z), bounds=(phi1 + 1e-7, phi2 - 1e-7), method="bounded",
                          options={"xatol": 1e-12})
    return -res.fun, (phi1, phi2, T, a, res.x)

# Find the Goode Homolosine junction latitude by solving p_moll(phi)=1.
junction_phi = brentq(
    lambda x: float(p_moll(x)) - 1.0,
    np.deg2rad(30),
    np.deg2rad(50),
    xtol=1e-13
)

def F_critical(phi2):
    return A2_max_for_phi2(phi2)[0]

# Automatically find a bracket for F_critical(phi2)=0.
def find_bracket_auto(lo, hi, n=800):
    xs = np.linspace(lo, hi, n)
    prev_x = None
    prev_y = None

    for x in xs:
        try:
            y = F_critical(x)
        except Exception:
            continue

        if not np.isfinite(y):
            continue

        if prev_y is not None:
            if prev_y > 0 and y <= 0:
                return prev_x, x

        prev_x = x
        prev_y = y

    raise RuntimeError("No sign-change bracket found.")

bracket = find_bracket_auto(
    junction_phi + 1e-7,
    np.pi / 2 - 1e-5
)

print("bracket deg:", np.rad2deg(bracket[0]), np.rad2deg(bracket[1]))

root_phi2 = brentq(
    F_critical,
    bracket[0],
    bracket[1],
    xtol=1e-11
)

maxA2, par = A2_max_for_phi2(root_phi2)
phi1, phi2, T, a, touch = par

print("bracket deg:", None if bracket is None else (np.rad2deg(bracket[0]), np.rad2deg(bracket[1])))

root_phi2 = brentq(lambda x: A2_max_for_phi2(x)[0], bracket[0], bracket[1], xtol=1e-11)
maxA2, par = A2_max_for_phi2(root_phi2)
phi1, phi2, T, a, touch = par

print("critical upper phi2 deg =", np.rad2deg(phi2))
print("critical lower phi1 deg =", np.rad2deg(phi1))
print("transition width T deg =", np.rad2deg(T))
print("a =", a)
print("touch latitude deg =", np.rad2deg(touch))
print("max A'' =", maxA2)
print("p(t)=1-a t^2; t in radians")

# Check sample A'' min/max across interval
xs = np.linspace(phi1, phi2, 3000)
def A_quad(x):
    t = x - phi1
    return np.cos(x)/(1-a*t*t)
A = A_quad(xs)
A2_arr = np.gradient(np.gradient(A, xs), xs)
print("sample A'' min/max:", A2_arr.min(), A2_arr.max())

# Plot p and A''
lat_deg = np.linspace(np.rad2deg(phi1)-3, np.rad2deg(phi2)+3, 2000)
ph = np.deg2rad(lat_deg)
p_quad = np.full_like(ph, np.nan, dtype=float)
mask = (ph >= phi1) & (ph <= phi2)
p_quad[mask] = 1 - a*(ph[mask]-phi1)**2

plt.figure(figsize=(9,5.4))
plt.plot(lat_deg, p_moll(ph), label="Mollweide p(φ)")
plt.plot(lat_deg, np.ones_like(ph), label="Sinusoidal p=1")
plt.plot(lat_deg, p_quad, linewidth=2.6, label=r"Critical quadratic transition $1-at^2$")
plt.axvline(np.rad2deg(phi1), linestyle="--", label=f"φ1≈{np.rad2deg(phi1):.4f}°")
plt.axvline(np.rad2deg(phi2), linestyle="--", label=f"φ2≈{np.rad2deg(phi2):.4f}°")
plt.xlabel("Latitude (degrees)")
plt.ylabel("Parallel spacing p(φ)")
plt.title("First concavity-preserving C¹ quadratic transition")
plt.grid(True)
plt.legend()
plt.tight_layout()
p_path = "/mnt/data/critical_quadratic_c1_spacing.png"
plt.savefig(p_path, dpi=220)
plt.show()

plt.figure(figsize=(9,5.2))
plt.plot(np.rad2deg(xs), A2_arr, linewidth=2)
plt.axhline(0, linewidth=1)
plt.axvline(np.rad2deg(touch), linestyle=":", label=f"touch≈{np.rad2deg(touch):.4f}°")
plt.xlabel("Latitude (degrees)")
plt.ylabel("A''(φ), numerical")
plt.title(r"Concavity boundary for $A(\phi)=\cos\phi/p(\phi)$")
plt.grid(True)
plt.legend()
plt.tight_layout()
a2_path = "/mnt/data/critical_quadratic_c1_A_second_derivative.png"
plt.savefig(a2_path, dpi=220)
plt.show()

p_path, a2_path
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq
from scipy.interpolate import PchipInterpolator
from scipy.integrate import cumulative_trapezoid
from matplotlib.collections import LineCollection
import geopandas as gpd
from shapely.geometry import shape, Polygon, MultiPolygon, LineString, MultiLineString, GeometryCollection

# -----------------------------
# 0. Critical quadratic C1 transition parameters
# -----------------------------

phi1_deg = 29.474562561747568
phi2_deg = 63.14381755893585
a = 0.5000000702370315

phi1 = np.deg2rad(phi1_deg)
phi2 = np.deg2rad(phi2_deg)

print(f"phi1 = {phi1_deg:.10f}°")
print(f"phi2 = {phi2_deg:.10f}°")
print(f"a = {a:.15g}")
print("p(t) = 1 - a*t^2, t in radians")


# -----------------------------
# 1. Mollweide parallel spacing and derivatives
# -----------------------------

def theta_from_phi(phi):
    phi = np.asarray(phi, dtype=float)
    theta = phi / 2
    for _ in range(70):
        f = 2*theta + np.sin(2*theta) - np.pi*np.sin(phi)
        fp = 2 + 2*np.cos(2*theta)
        theta = np.where(np.abs(fp) > 1e-12, theta - f/fp, theta)
    return theta

def p_moll(phi):
    th = theta_from_phi(phi)
    return np.pi*np.cos(phi)/(2*np.sqrt(2)*np.cos(th))

def derivs_p_moll(phi):
    h = 1e-5
    P = p_moll(phi)
    Q = (p_moll(phi+h) - p_moll(phi-h))/(2*h)
    R = (p_moll(phi+h) - 2*p_moll(phi) + p_moll(phi-h))/(h*h)
    return P, Q, R

# -----------------------------
# 2. Solve phi2 and quartic p(t)=1-a*t^3-b*t^4
#    Matching p, p', p'' to Mollweide at phi2
# -----------------------------

def ab_from(phi1, phi2):
    T = phi2 - phi1
    P, Q, R = derivs_p_moll(phi2)
    M = np.array([[3*T*T, 4*T**3], [6*T, 12*T*T]], dtype=float)
    rhs = np.array([-Q, -R], dtype=float)
    a, b = np.linalg.solve(M, rhs)
    return a, b

def residual_phi2(phi2, phi1):
    T = phi2 - phi1
    if T <= 0:
        return np.nan
    P, Q, R = derivs_p_moll(phi2)
    a, b = ab_from(phi1, phi2)
    return 1 - a*T**3 - b*T**4 - P

def find_phi2(phi1):
    xs = np.deg2rad(np.linspace(np.rad2deg(phi1)+0.2, 88.5, 700))
    vals = np.array([residual_phi2(x, phi1) for x in xs])
    for i in range(len(xs)-1):
        if np.isfinite(vals[i]) and np.isfinite(vals[i+1]) and vals[i]*vals[i+1] < 0:
            return brentq(lambda z: residual_phi2(z, phi1), xs[i], xs[i+1], xtol=1e-13)
    raise RuntimeError("No phi2 root found")

phi1_deg = 29.474562561747568
phi2_deg = 63.14381755893585
a = 0.5000000702370315

print(f"phi1 = {phi1_deg:.10f}°")
print(f"phi2 = {phi2_deg:.10f}°")
print(f"T = {phi2_deg - phi1_deg:.10f}°")
print(f"a = {a:.15g}")
print(f"p(t) = 1 - ({a:.15g}) t^2 , t in radians")

# -----------------------------
# 3. Plot parallel spacing
# -----------------------------

lat_deg = np.linspace(-90, 90, 6001)
phi = np.deg2rad(lat_deg)
u = np.abs(phi)
t = u - phi1
p_trans = 1 - a*t*t
p_piece = np.where(u <= phi1, 1.0, np.where(u <= phi2, p_trans, p_moll(phi)))




plt.figure(figsize=(9.5,5.7))
plt.plot(lat_deg, np.ones_like(lat_deg), label="Sinusoidal: 1", linewidth=1.3)
plt.plot(lat_deg, p_moll(phi), label="Mollweide", linewidth=1.3)
plt.plot(lat_deg, p_piece, label="Critical C¹ quadratic transition", linewidth=2.8)
for d in [-phi2_deg, -phi1_deg, phi1_deg, phi2_deg]:
    plt.axvline(d, linestyle="--", linewidth=1)
plt.xlabel("Latitude (degrees)")
plt.ylabel("Parallel spacing function dy/dφ")
plt.title("Parallel spacing with critical C¹ quadratic transition")
plt.grid(True)
plt.legend()
plt.tight_layout()
spacing_path = "/mnt/data/parallel_spacing_critical_c1_quadratic_transition.png"
plt.savefig(spacing_path, dpi=220)
plt.show()

# -----------------------------
# 4. Build projection from p(phi): y = ∫p dphi, A = cos(phi)/p
# -----------------------------
def p_spacing(phi):
    phi = np.asarray(phi, dtype=float)
    u = np.abs(phi)
    t = u - phi1
    pt = 1 - a*t*t
    return np.where(u <= phi1, 1.0, np.where(u <= phi2, pt, p_moll(phi)))
    
grid_phi = np.linspace(0, np.pi/2 - 1e-7, 30000)
grid_y = cumulative_trapezoid(p_spacing(grid_phi), grid_phi, initial=0)
y_interp = PchipInterpolator(grid_phi, grid_y, extrapolate=True)

def y_of_phi(phi):
    return np.sign(phi)*y_interp(np.abs(phi))

def A_of_phi(phi):
    return np.cos(phi)/p_spacing(phi)

def project_lonlat(lon_deg, lat_deg):
    lon = np.deg2rad(lon_deg)
    ph = np.deg2rad(lat_deg)
    return lon*A_of_phi(ph), y_of_phi(ph)

# -----------------------------
# 5. Geometry helpers and world map
# -----------------------------

def split_ring(coords):
    if len(coords) < 2: 
        return []
    segs, cur = [], [coords[0]]
    prev = coords[0][0]
    for pt in coords[1:]:
        lon = pt[0]
        if abs(lon-prev) > 180:
            if len(cur) >= 2:
                segs.append(cur)
            cur = [pt]
        else:
            cur.append(pt)
        prev = lon
    if len(cur) >= 2:
        segs.append(cur)
    return segs

def densify(coords, max_step=0.75):
    out = []
    for (lon1,lat1),(lon2,lat2) in zip(coords[:-1], coords[1:]):
        out.append((lon1,lat1))
        n = int(max(abs(lon2-lon1), abs(lat2-lat1)) // max_step)
        for k in range(1,n+1):
            f = k/(n+1)
            out.append((lon1+f*(lon2-lon1), lat1+f*(lat2-lat1)))
    out.append(coords[-1])
    return out

def geom_segments(geom):
    segs = []
    if isinstance(geom, Polygon):
        rings = [geom.exterior] + list(geom.interiors)
        for r in rings:
            for part in split_ring(list(r.coords)):
                d = densify(part)
                lon = np.array([p[0] for p in d])
                lat = np.array([p[1] for p in d])
                x,y = project_lonlat(lon, lat)
                segs.append(np.column_stack([x,y]))
    elif isinstance(geom, MultiPolygon):
        for g in geom.geoms:
            segs.extend(geom_segments(g))
    elif isinstance(geom, LineString):
        for part in split_ring(list(geom.coords)):
            d = densify(part)
            lon = np.array([p[0] for p in d])
            lat = np.array([p[1] for p in d])
            x,y = project_lonlat(lon, lat)
            segs.append(np.column_stack([x,y]))
    elif isinstance(geom, MultiLineString):
        for g in geom.geoms:
            segs.extend(geom_segments(g))
    elif isinstance(geom, GeometryCollection):
        for g in geom.geoms:
            segs.extend(geom_segments(g))
    return segs

#world = gpd.read_file(shp_path)
url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
world = gpd.read_file(url)
coast_segments = []
for geom in world.geometry:
    coast_segments.extend(geom_segments(geom))

grat = []
for lon in range(-180, 181, 30):
    lats = np.linspace(-89.999, 89.999, 800)
    lons = np.full_like(lats, lon)
    x,y = project_lonlat(lons,lats)
    grat.append(np.column_stack([x,y]))

for lat in range(-75, 76, 15):
    lons = np.linspace(-180, 180, 1200)
    lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats)
    grat.append(np.column_stack([x,y]))

boundary = []
for lat in [89.999, -89.999]:
    lons = np.linspace(-180,180,1200)
    lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats)
    boundary.append(np.column_stack([x,y]))

for lon in [-180,180]:
    lats = np.linspace(-89.999,89.999,1200)
    lons = np.full_like(lats,lon)
    x,y = project_lonlat(lons,lats)
    boundary.append(np.column_stack([x,y]))

transition_lines = []
for lat in [phi1_deg, -phi1_deg, phi2_deg, -phi2_deg]:
    lons = np.linspace(-180,180,1200)
    lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats)
    transition_lines.append(np.column_stack([x,y]))

fig, ax = plt.subplots(figsize=(13.5,7.2))
ax.add_collection(LineCollection(grat, linewidths=0.45, alpha=0.45))
ax.add_collection(LineCollection(coast_segments, linewidths=0.6))
ax.add_collection(LineCollection(boundary, linewidths=1.0))
ax.add_collection(LineCollection(transition_lines, linewidths=0.95, linestyles="--"))

ax.set_aspect("equal", adjustable="box")
ax.autoscale()
ax.set_title("Critical C¹ quadratic-transition equal-area world map")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.text(
    0.01, 0.02,
    f"phi1≈{phi1_deg:.5f}° (Tropic), phi2≈{phi2_deg:.5f}°",
    transform=ax.transAxes,
    fontsize=9,
    va="bottom"
)
plt.tight_layout()
map_path = "/mnt/data/critical_c1_quadratic_transition_world_map.png"
plt.savefig(map_path, dpi=250)
plt.show()

spacing_path, map_path
```

### Polynomial Transition Without the Mollweide Region

For comparison, we temporarily remove the Mollweide portion of the projection and extend the polynomial transition all the way to the poles. This construction is not intended as a practical projection; rather, it isolates the geometric effect of the polynomial transition itself.

By comparing the resulting map with the standard Sinusoidal projection, we can more clearly observe how the transition function modifies meridian shape, parallel spacing, and overall map appearance, independent of the Mollweide component.


$
p(\phi)=1-a t^3-b t^4\qquad t=\phi-\phi_1.
$


Although the transition function contains three parameters, \(a\), \(b\), and \(\phi_1\), the lower junction conditions at \(t=0\) are satisfied identically

$
p(\phi_1)=1,
p'(\phi_1)=0,
p''(\phi_1)=0.
$



Thus the Sinusoidal–polynomial matching requires no additional constraints.

At the upper junction latitude \(\phi_2\), the polynomial is required to match the Mollweide spacing function and its first two derivatives:

$
p(\phi_2)=p_M(\phi_2),
p'(\phi_2)=p_M'(\phi_2),
p''(\phi_2)=p_M''(\phi_2).
$


These three conditions determine \(a\), \(b\), and \(\phi_1\) uniquely for a given choice of \(\phi_2\).

Therefore, the admissible tuples \((a,b,\phi_1,\phi_2)\) form a one-parameter family. Equivalently, the admissible pairs \((\phi_1,\phi_2)\) lie on a one-dimensional curve in parameter space, so the transition family possesses only one effective degree of freedom.

In the quartic case,
$p(\phi)=1-a t^3 -b t^4$


the parameters \(a\), \(b\), and \(\phi_1\) are determined by the three upper-junction matching conditions for any prescribed value of \(\phi_2\). Consequently, the family retains one effective degree of freedom, represented by a continuous range of admissible transition latitudes.

By contrast, if the quartic term is removed,

$
p(\phi)=1-a t^3,
$

the number of free parameters is reduced by one. The upper-junction \(C^2\) matching conditions then completely determine \(a\), \(\phi_1\), and \(\phi_2\). As a result, the family possesses no remaining degree of freedom. At most a single admissible transition interval can exist.

```python
import mpmath as mp

mp.mp.dps = 80

def theta_from_phi(phi):
    f = lambda th: 2*th + mp.sin(2*th) - mp.pi*mp.sin(phi)
    return mp.findroot(f, phi/2)

def p_moll(phi):
    th = theta_from_phi(phi)
    return mp.pi * mp.cos(phi) / (2 * mp.sqrt(2) * mp.cos(th))

def dp_moll(phi):
    return mp.diff(p_moll, phi, 1)

def ddp_moll(phi):
    return mp.diff(p_moll, phi, 2)

# For p(t)=1-a t^3:
# P = 1 - aT^3
# Q = -3aT^2
# R = -6aT
#
# Eliminating a and T gives:
# R = -2 Q^2 / (3(1-P))
# Define residual:
def cubic_condition_residual(phi):
    P = p_moll(phi)
    Q = dp_moll(phi)
    R = ddp_moll(phi)
    return R + 2*Q**2/(3*(1-P))

# Search for sign changes over valid upper transition latitudes.
# Need P < 1 so that 1-P > 0; this is above the usual sinu-Mollweide crossing.
samples = []
for deg in [40 + i*0.1 for i in range(1, 500)]:  # 40.1 to 89.9
    phi = mp.radians(deg)
    try:
        res = cubic_condition_residual(phi)
        samples.append((phi, res))
    except Exception:
        pass

roots = []
for (a, fa), (b, fb) in zip(samples[:-1], samples[1:]):
    if fa == 0 or fa*fb < 0:
        try:
            r = mp.findroot(cubic_condition_residual, (a, b))
            if 0 < r < mp.pi/2 and all(abs(r-x) > mp.mpf("1e-30") for x in roots):
                roots.append(r)
        except Exception:
            pass

print("number of roots:", len(roots))

for phi2 in roots:
    P = p_moll(phi2)
    Q = dp_moll(phi2)
    R = ddp_moll(phi2)
    
    # From Q/R = T/2 for cubic
    T = 2*Q/R
    a = (1-P)/T**3
    phi1 = phi2 - T
    
    print("\nphi2 deg =", mp.degrees(phi2))
    print("P =", P)
    print("Q =", Q)
    print("R =", R)
    print("T rad =", T)
    print("T deg =", mp.degrees(T))
    print("phi1 deg =", mp.degrees(phi1))
    print("a =", a)
    print("check P =", 1 - a*T**3)
    print("check Q =", -3*a*T**2)
    print("check R =", -6*a*T)

# Also print residual min rough info for context
min_abs = min(samples, key=lambda x: abs(x[1]))
print("\nrough smallest residual sample:")
print("deg =", mp.degrees(min_abs[0]))
print("residual =", min_abs[1])
```

### Quartic with a Shorter Transition Interval

The quartic transition function permits a lower transition latitude \(\phi_1\) than the \(32.28^\circ\) used by the \(C^1\)-smooth Goode Homolosine projection.

Nevertheless, reducing \(\phi_1\) only slightly below this value is insufficient to obtain a convex map. The resulting projection still violates the desired concavity condition, demonstrating that the available transition interval is too short despite the additional degree of freedom introduced by the quartic term.

```python
import mpmath as mp
import numpy as np
import matplotlib.pyplot as plt

mp.mp.dps = 80

# -----------------------------
# Mollweide parallel spacing p(phi)
# -----------------------------

def theta_from_phi(phi):
    f = lambda th: 2*th + mp.sin(2*th) - mp.pi*mp.sin(phi)
    return mp.findroot(f, phi/2)

def p_moll(phi):
    th = theta_from_phi(phi)
    return mp.pi * mp.cos(phi) / (2 * mp.sqrt(2) * mp.cos(th))

def dp_moll(phi):
    return mp.diff(p_moll, phi, 1)

def ddp_moll(phi):
    return mp.diff(p_moll, phi, 2)

# -----------------------------
# Transition function:
# p(t) = 1 - a t^3 - b t^4, t = phi - phi1
#
# At t = T = phi2 - phi1:
# 1 - aT^3 - bT^4 = P
# -3aT^2 - 4bT^3 = Q
# -6aT - 12bT^2 = R
#
# For a given phi1 and phi2, use Q,R to solve a,b,
# then impose the value condition as one scalar residual.
# -----------------------------

phi1 = mp.radians(30)

def ab_from_phi2(phi2):
    T = phi2 - phi1
    Q = dp_moll(phi2)
    R = ddp_moll(phi2)
    
    # Solve:
    # 3aT^2 + 4bT^3 = -Q
    # 6aT  +12bT^2 = -R
    det = (3*T**2)*(12*T**2) - (4*T**3)*(6*T)
    a = ((-Q)*(12*T**2) - (4*T**3)*(-R)) / det
    b = ((3*T**2)*(-R) - (-Q)*(6*T)) / det
    return a, b

def residual_phi2(phi2):
    T = phi2 - phi1
    if T <= 0:
        return mp.nan
    P = p_moll(phi2)
    a, b = ab_from_phi2(phi2)
    return 1 - a*T**3 - b*T**4 - P

# Search phi2 > phi1
samples = []
for deg in np.linspace(30.1, 89.5, 1000):
    ph = mp.radians(float(deg))
    try:
        res = residual_phi2(ph)
        if mp.isfinite(res):
            samples.append((ph, res))
    except Exception:
        pass

roots = []
for (x0, y0), (x1, y1) in zip(samples[:-1], samples[1:]):
    try:
        if y0 == 0 or y0*y1 < 0:
            r = mp.findroot(residual_phi2, (x0, x1))
            if phi1 < r < mp.pi/2 and all(abs(r - old) > mp.mpf("1e-30") for old in roots):
                roots.append(r)
    except Exception:
        pass

print("roots found:", len(roots))

for phi2 in roots:
    T = phi2 - phi1
    P = p_moll(phi2)
    Q = dp_moll(phi2)
    R = ddp_moll(phi2)
    a, b = ab_from_phi2(phi2)
    
    print("\nphi1 =", mp.degrees(phi1), "deg")
    print("phi2 =", mp.degrees(phi2), "deg")
    print("T =", T, "rad")
    print("T =", mp.degrees(T), "deg")
    print("P =", P)
    print("Q =", Q)
    print("R =", R)
    print("a =", a)
    print("b =", b)
    print("value check =", 1 - a*T**3 - b*T**4)
    print("first derivative check =", -3*a*T**2 - 4*b*T**3)
    print("second derivative check =", -6*a*T - 12*b*T**2)

# Pick first valid root
phi2 = roots[0]
T = phi2 - phi1
a_mp, b_mp = ab_from_phi2(phi2)

# -----------------------------
# Plot spacing functions
# -----------------------------

phi1_f = float(phi1)
phi2_f = float(phi2)
a_f = float(a_mp)
b_f = float(b_mp)

lat_deg = np.linspace(-89.999, 89.999, 5000)
phi = np.deg2rad(lat_deg)
u = np.abs(phi)

# Numpy Mollweide spacing
def theta_from_phi_np(phi_arr):
    phi_arr = np.asarray(phi_arr, dtype=float)
    theta = phi_arr / 2
    for _ in range(80):
        f = 2*theta + np.sin(2*theta) - np.pi*np.sin(phi_arr)
        fp = 2 + 2*np.cos(2*theta)
        theta = np.where(np.abs(fp) > 1e-12, theta - f/fp, theta)
    return theta

theta = theta_from_phi_np(phi)
p_moll_np = np.pi * np.cos(phi) / (2*np.sqrt(2)*np.cos(theta))

p_sinu = np.ones_like(phi)
t = u - phi1_f
p_transition = 1 - a_f*t**3 - b_f*t**4

p_piecewise = np.where(
    u <= phi1_f,
    p_sinu,
    np.where(u <= phi2_f, p_transition, p_moll_np)
)

plt.figure(figsize=(9.5, 5.7))
plt.plot(lat_deg, p_sinu, label="Sinusoidal: 1", linewidth=1.3)
plt.plot(lat_deg, p_moll_np, label="Mollweide", linewidth=1.3)
plt.plot(lat_deg, p_piecewise, label=r"C² quartic transition: $1-at^3-bt^4$", linewidth=2.8)

for d in [-np.rad2deg(phi2_f), -np.rad2deg(phi1_f), np.rad2deg(phi1_f), np.rad2deg(phi2_f)]:
    plt.axvline(d, linestyle="--", linewidth=1)

plt.xlabel("Latitude (degrees)")
plt.ylabel("Parallel spacing function dy/dφ")
plt.title(r"Parallel spacing function, $\phi_1=30^\circ$, quartic transition")
plt.grid(True)
plt.legend()
plt.tight_layout()

path = "/mnt/data/parallel_spacing_quartic_transition_phi1_30.png"
plt.savefig(path, dpi=220)
plt.show()

print("\nplot path:", path)
print("formula:")
print(f"p(t) = 1 - ({a_f:.15g}) t^3 - ({b_f:.15g}) t^4,  t in radians")
print("phi1_deg:", float(mp.degrees(phi1)))
print("phi2_deg:", float(mp.degrees(phi2)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import fiona
from shapely.geometry import shape, Polygon, MultiPolygon, LineString, MultiLineString, GeometryCollection
from scipy.interpolate import PchipInterpolator
from scipy.integrate import cumulative_trapezoid

# -----------------------------
# Projection parameters
# -----------------------------

phi1_deg = 30.0
phi2_deg = 52.61997546696894

# Transition function:
# p(t)=1-a t^3 + c t^4
# t in radians
a = 2.2840341768984023
c = 2.5395240596949243

phi1 = np.deg2rad(phi1_deg)
phi2 = np.deg2rad(phi2_deg)

# -----------------------------
# Mollweide functions
# -----------------------------

def theta_from_phi(phi_arr):
    phi_arr = np.asarray(phi_arr, dtype=float)
    theta = phi_arr / 2
    for _ in range(80):
        f = 2*theta + np.sin(2*theta) - np.pi*np.sin(phi_arr)
        fp = 2 + 2*np.cos(2*theta)
        theta = np.where(np.abs(fp) > 1e-12, theta - f/fp, theta)
    return theta

def p_moll(phi):
    th = theta_from_phi(phi)
    return np.pi * np.cos(phi) / (2*np.sqrt(2)*np.cos(th))

# -----------------------------
# Parallel spacing p(phi)=dy/dphi
# -----------------------------

def p_spacing(phi):
    phi = np.asarray(phi, dtype=float)
    u = np.abs(phi)
    t = u - phi1
    
    p_transition = 1 - a*t**3 + c*t**4
    
    return np.where(
        u <= phi1,
        1.0,
        np.where(u <= phi2, p_transition, p_moll(phi))
    )

# -----------------------------
# Build y(phi) by integration
# -----------------------------

grid_phi = np.linspace(0, np.pi/2 - 1e-7, 40000)
grid_p = p_spacing(grid_phi)
grid_y = cumulative_trapezoid(grid_p, grid_phi, initial=0.0)

y_interp = PchipInterpolator(grid_phi, grid_y, extrapolate=True)

def y_of_phi(phi):
    phi = np.asarray(phi, dtype=float)
    return np.sign(phi) * y_interp(np.abs(phi))

# Equal-area condition:
# A(phi)*p(phi)=cos(phi)
def A_of_phi(phi):
    phi = np.asarray(phi, dtype=float)
    return np.cos(phi) / p_spacing(phi)

def project_lonlat(lon_deg, lat_deg):
    lon = np.deg2rad(lon_deg)
    phi = np.deg2rad(lat_deg)
    x = lon * A_of_phi(phi)
    y = y_of_phi(phi)
    return x, y

# -----------------------------
# Geometry helpers
# -----------------------------

def split_ring_at_dateline(coords):
    if len(coords) < 2:
        return []
    segments = []
    current = [coords[0]]
    prev_lon = coords[0][0]
    for pt in coords[1:]:
        lon = pt[0]
        if abs(lon - prev_lon) > 180:
            if len(current) >= 2:
                segments.append(current)
            current = [pt]
        else:
            current.append(pt)
        prev_lon = lon
    if len(current) >= 2:
        segments.append(current)
    return segments

def densify_coords(coords, max_step=0.5):
    out = []
    for (lon1, lat1), (lon2, lat2) in zip(coords[:-1], coords[1:]):
        out.append((lon1, lat1))
        n = int(max(abs(lon2-lon1), abs(lat2-lat1)) // max_step)
        for k in range(1, n+1):
            f = k/(n+1)
            out.append((lon1 + f*(lon2-lon1), lat1 + f*(lat2-lat1)))
    out.append(coords[-1])
    return out

def projected_segments_from_geom(geom):
    segments = []
    
    if isinstance(geom, Polygon):
        rings = [geom.exterior] + list(geom.interiors)
        for ring in rings:
            coords = list(ring.coords)
            for part in split_ring_at_dateline(coords):
                dense = densify_coords(part)
                lon = np.array([p[0] for p in dense])
                lat = np.array([p[1] for p in dense])
                x, y = project_lonlat(lon, lat)
                segments.append(np.column_stack([x, y]))
                
    elif isinstance(geom, MultiPolygon):
        for g in geom.geoms:
            segments.extend(projected_segments_from_geom(g))
            
    elif isinstance(geom, LineString):
        coords = list(geom.coords)
        for part in split_ring_at_dateline(coords):
            dense = densify_coords(part)
            lon = np.array([p[0] for p in dense])
            lat = np.array([p[1] for p in dense])
            x, y = project_lonlat(lon, lat)
            segments.append(np.column_stack([x, y]))
            
    elif isinstance(geom, MultiLineString):
        for g in geom.geoms:
            segments.extend(projected_segments_from_geom(g))
            
    elif isinstance(geom, GeometryCollection):
        for g in geom.geoms:
            segments.extend(projected_segments_from_geom(g))
            
    return segments

# -----------------------------
# Load Natural Earth
# -----------------------------


#world = gpd.read_file(shp_path)
url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
world = gpd.read_file(url)
coast_segments = []
for geom in world.geometry:
    coast_segments.extend(projected_segments_from_geom(geom))


# -----------------------------
# Graticule
# -----------------------------

grat_segments = []

# Meridians
for lon in range(-180, 181, 30):
    lats = np.linspace(-89.999, 89.999, 900)
    lons = np.full_like(lats, lon)
    x, y = project_lonlat(lons, lats)
    grat_segments.append(np.column_stack([x, y]))

# Parallels
for lat in range(-75, 76, 15):
    lons = np.linspace(-180, 180, 1400)
    lats = np.full_like(lons, lat)
    x, y = project_lonlat(lons, lats)
    grat_segments.append(np.column_stack([x, y]))

# Boundary
boundary_segments = []

for lat in [89.999, -89.999]:
    lons = np.linspace(-180, 180, 1500)
    lats = np.full_like(lons, lat)
    x, y = project_lonlat(lons, lats)
    boundary_segments.append(np.column_stack([x, y]))

for lon in [-180, 180]:
    lats = np.linspace(-89.999, 89.999, 1500)
    lons = np.full_like(lats, lon)
    x, y = project_lonlat(lons, lats)
    boundary_segments.append(np.column_stack([x, y]))

# Transition latitude lines
transition_segments = []

for lat in [phi1_deg, -phi1_deg, phi2_deg, -phi2_deg]:
    lons = np.linspace(-180, 180, 1400)
    lats = np.full_like(lons, lat)
    x, y = project_lonlat(lons, lats)
    transition_segments.append(np.column_stack([x, y]))

# -----------------------------
# Plot
# -----------------------------

fig, ax = plt.subplots(figsize=(13.5, 7.2))

ax.add_collection(LineCollection(grat_segments, linewidths=0.45, alpha=0.45))
ax.add_collection(LineCollection(coast_segments, linewidths=0.6))
ax.add_collection(LineCollection(boundary_segments, linewidths=1.0))
ax.add_collection(LineCollection(transition_segments, linewidths=0.9, linestyles="--"))

ax.set_aspect("equal", adjustable="box")
ax.autoscale()

ax.set_title(r"Equal-area world map with $C^2$ quartic transition")
ax.set_xlabel("x")
ax.set_ylabel("y")

ax.text(
    0.01, 0.02,
    r"$p(t)=1-at^3+ct^4,\ \phi_1=30^\circ,\ \phi_2\approx52.62^\circ$",
    transform=ax.transAxes,
    fontsize=9,
    va="bottom"
)

plt.tight_layout()

path = "/mnt/data/c2_quartic_transition_world_map_phi1_30.png"
plt.savefig(path, dpi=250)
plt.show()

path
```

When \(\phi_1\) is close to the Kirk junction latitude used by the Goode Homolosine projection, the transition yields a strongly concave map boundary. As \(\phi_1\) decreases, this concavity becomes progressively weaker.

It is therefore natural to conjecture that a critical value of \(\phi_1\) exists at which the boundary curvature becomes zero. For smaller values of \(\phi_1\), the boundary is expected to become convex.

To identify the shortest admissible transition interval, we search for the value of \(\phi_1\) corresponding to this curvature-zero boundary case.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq, minimize_scalar
from scipy.interpolate import PchipInterpolator
from scipy.integrate import cumulative_trapezoid
from matplotlib.collections import LineCollection
import fiona
from shapely.geometry import shape, Polygon, MultiPolygon, LineString, MultiLineString, GeometryCollection

# -----------------------------
# Fast Mollweide p(phi) and numeric derivatives
# -----------------------------

def theta_from_phi(phi):
    phi = np.asarray(phi, dtype=float)
    theta = phi / 2
    for _ in range(60):
        f = 2*theta + np.sin(2*theta) - np.pi*np.sin(phi)
        fp = 2 + 2*np.cos(2*theta)
        theta = np.where(np.abs(fp) > 1e-12, theta - f/fp, theta)
    return theta

def p_moll(phi):
    th = theta_from_phi(phi)
    return np.pi*np.cos(phi)/(2*np.sqrt(2)*np.cos(th))

def derivs_p_moll(phi):
    h = 1e-5
    P = p_moll(phi)
    Q = (p_moll(phi+h) - p_moll(phi-h))/(2*h)
    R = (p_moll(phi+h) - 2*p_moll(phi) + p_moll(phi-h))/(h*h)
    return P, Q, R

def ab_from(phi1, phi2):
    T = phi2 - phi1
    P, Q, R = derivs_p_moll(phi2)
    # 3aT^2 + 4bT^3 = -Q
    # 6aT   +12bT^2 = -R
    M = np.array([[3*T*T, 4*T**3], [6*T, 12*T*T]], dtype=float)
    rhs = np.array([-Q, -R], dtype=float)
    a, b = np.linalg.solve(M, rhs)
    return a, b

def residual_phi2(phi2, phi1):
    T = phi2 - phi1
    if T <= 0:
        return np.nan
    P, Q, R = derivs_p_moll(phi2)
    a, b = ab_from(phi1, phi2)
    return 1 - a*T**3 - b*T**4 - P

def find_phi2(phi1):
    xs = np.deg2rad(np.linspace(np.rad2deg(phi1)+0.2, 88.5, 500))
    vals = []
    for x in xs:
        try:
            vals.append(residual_phi2(x, phi1))
        except Exception:
            vals.append(np.nan)
    vals = np.array(vals)
    for i in range(len(xs)-1):
        if np.isfinite(vals[i]) and np.isfinite(vals[i+1]) and vals[i]*vals[i+1] < 0:
            return brentq(lambda z: residual_phi2(z, phi1), xs[i], xs[i+1], xtol=1e-12)
    raise RuntimeError("no phi2")

def max_A2(phi1):
    phi2 = find_phi2(phi1)
    a, b = ab_from(phi1, phi2)
    
    def A(x):
        t = x - phi1
        return np.cos(x)/(1 - a*t**3 - b*t**4)
    
    def A2_num(x):
        h = 2e-4
        return (A(x+h) - 2*A(x) + A(x-h))/(h*h)
    
    res = minimize_scalar(lambda z: -A2_num(z), bounds=(phi1+1e-5, phi2-1e-5), method="bounded",
                          options={"xatol":1e-10})
    return -res.fun, res.x, phi2, a, b

def boundary_residual(phi1):
    return max_A2(phi1)[0]

# Find boundary phi1 where max A'' = 0
lo, hi = np.deg2rad(25), np.deg2rad(30)
# scan bracket
xs = np.deg2rad(np.linspace(25, 30, 26))
prev_x, prev_v = xs[0], boundary_residual(xs[0])
bracket = None
for x in xs[1:]:
    v = boundary_residual(x)
    if prev_v * v < 0:
        bracket = (prev_x, x)
        break
    prev_x, prev_v = x, v

phi1_star = brentq(lambda z: boundary_residual(z), bracket[0], bracket[1], xtol=1e-10)
mx, touch, phi2_star, a_star, b_star = max_A2(phi1_star)

print("phi1* deg =", np.rad2deg(phi1_star))
print("phi2* deg =", np.rad2deg(phi2_star))
print("touch latitude deg =", np.rad2deg(touch))
print("max A'' ≈", mx)
print("a =", a_star)
print("b =", b_star)
print("p(t)=1-a t^3-b t^4, t in radians")

# -----------------------------
# Plot parallel spacing
# -----------------------------
lat_deg = np.linspace(-89.999, 89.999, 5000)
phi = np.deg2rad(lat_deg)
u = np.abs(phi)
t = u - phi1_star
p_trans = 1 - a_star*t**3 - b_star*t**4
p_piece = np.where(u <= phi1_star, 1.0, np.where(u <= phi2_star, p_trans, p_moll(phi)))

plt.figure(figsize=(9.5,5.7))
plt.plot(lat_deg, np.ones_like(lat_deg), label="Sinusoidal: 1", linewidth=1.3)
plt.plot(lat_deg, p_moll(phi), label="Mollweide", linewidth=1.3)
plt.plot(lat_deg, p_piece, label="Quartic boundary map", linewidth=2.8)
for d in [-np.rad2deg(phi2_star), -np.rad2deg(phi1_star), np.rad2deg(phi1_star), np.rad2deg(phi2_star)]:
    plt.axvline(d, linestyle="--", linewidth=1)
for d in [-np.rad2deg(touch), np.rad2deg(touch)]:
    plt.axvline(d, linestyle=":", linewidth=1.2)
plt.xlabel("Latitude (degrees)")
plt.ylabel("Parallel spacing function dy/dφ")
plt.title("Parallel spacing at the concavity-boundary quartic transition")
plt.grid(True)
plt.legend()
plt.tight_layout()
spacing_path = "/mnt/data/parallel_spacing_quartic_concavity_boundary.png"
plt.savefig(spacing_path, dpi=220)
plt.show()

# -----------------------------
# World map
# -----------------------------
def p_spacing(phi):
    phi = np.asarray(phi, dtype=float)
    u = np.abs(phi)
    t = u - phi1_star
    pt = 1 - a_star*t**3 - b_star*t**4
    return np.where(u <= phi1_star, 1.0, np.where(u <= phi2_star, pt, p_moll(phi)))

grid_phi = np.linspace(0, np.pi/2 - 1e-7, 30000)
grid_y = cumulative_trapezoid(p_spacing(grid_phi), grid_phi, initial=0)
y_interp = PchipInterpolator(grid_phi, grid_y, extrapolate=True)

def y_of_phi(phi):
    return np.sign(phi)*y_interp(np.abs(phi))

def A_of_phi(phi):
    return np.cos(phi)/p_spacing(phi)

def project_lonlat(lon_deg, lat_deg):
    lon = np.deg2rad(lon_deg)
    ph = np.deg2rad(lat_deg)
    return lon*A_of_phi(ph), y_of_phi(ph)

def split_ring(coords):
    if len(coords) < 2: return []
    segs, cur = [], [coords[0]]
    prev = coords[0][0]
    for pt in coords[1:]:
        lon = pt[0]
        if abs(lon-prev) > 180:
            if len(cur) >= 2: segs.append(cur)
            cur = [pt]
        else:
            cur.append(pt)
        prev = lon
    if len(cur) >= 2: segs.append(cur)
    return segs

def densify(coords, max_step=0.75):
    out = []
    for (lon1,lat1),(lon2,lat2) in zip(coords[:-1], coords[1:]):
        out.append((lon1,lat1))
        n = int(max(abs(lon2-lon1), abs(lat2-lat1)) // max_step)
        for k in range(1,n+1):
            f = k/(n+1)
            out.append((lon1+f*(lon2-lon1), lat1+f*(lat2-lat1)))
    out.append(coords[-1])
    return out

def geom_segments(geom):
    segs = []
    if isinstance(geom, Polygon):
        rings = [geom.exterior] + list(geom.interiors)
        for r in rings:
            for part in split_ring(list(r.coords)):
                d = densify(part)
                lon = np.array([p[0] for p in d]); lat = np.array([p[1] for p in d])
                x,y = project_lonlat(lon, lat)
                segs.append(np.column_stack([x,y]))
    elif isinstance(geom, MultiPolygon):
        for g in geom.geoms: segs.extend(geom_segments(g))
    elif isinstance(geom, LineString):
        for part in split_ring(list(geom.coords)):
            d = densify(part)
            lon = np.array([p[0] for p in d]); lat = np.array([p[1] for p in d])
            x,y = project_lonlat(lon, lat)
            segs.append(np.column_stack([x,y]))
    elif isinstance(geom, MultiLineString):
        for g in geom.geoms: segs.extend(geom_segments(g))
    elif isinstance(geom, GeometryCollection):
        for g in geom.geoms: segs.extend(geom_segments(g))
    return segs


#world = gpd.read_file(shp_path)
url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
world = gpd.read_file(url)
coast_segments = []
for geom in world.geometry:
    coast_segments.extend(geom_segments(geom))

grat = []
for lon in range(-180, 181, 30):
    lats = np.linspace(-89.999, 89.999, 800); lons = np.full_like(lats, lon)
    x,y = project_lonlat(lons,lats); grat.append(np.column_stack([x,y]))
for lat in range(-75, 76, 15):
    lons = np.linspace(-180, 180, 1200); lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats); grat.append(np.column_stack([x,y]))

boundary = []
for lat in [89.999,-89.999]:
    lons = np.linspace(-180,180,1200); lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats); boundary.append(np.column_stack([x,y]))
for lon in [-180,180]:
    lats = np.linspace(-89.999,89.999,1200); lons = np.full_like(lats,lon)
    x,y = project_lonlat(lons,lats); boundary.append(np.column_stack([x,y]))

lines = []
for lat in [np.rad2deg(phi1_star), -np.rad2deg(phi1_star), np.rad2deg(phi2_star), -np.rad2deg(phi2_star)]:
    lons = np.linspace(-180,180,1200); lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats); lines.append(np.column_stack([x,y]))
touch_lines = []
for lat in [np.rad2deg(touch), -np.rad2deg(touch)]:
    lons = np.linspace(-180,180,1200); lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats); touch_lines.append(np.column_stack([x,y]))

fig, ax = plt.subplots(figsize=(13.5,7.2))
ax.add_collection(LineCollection(grat, linewidths=0.45, alpha=0.45))
ax.add_collection(LineCollection(coast_segments, linewidths=0.6))
ax.add_collection(LineCollection(boundary, linewidths=1.0))
ax.add_collection(LineCollection(lines, linewidths=0.9, linestyles="--"))
ax.add_collection(LineCollection(touch_lines, linewidths=0.9, linestyles=":"))
ax.set_aspect("equal", adjustable="box")
ax.autoscale()
ax.set_title("Equal-area world map at the parallel-length concavity boundary")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.text(0.01,0.02,
        f"phi1={np.rad2deg(phi1_star):.4f}°, phi2={np.rad2deg(phi2_star):.4f}°, A'' touch≈{np.rad2deg(touch):.4f}°",
        transform=ax.transAxes, fontsize=9, va="bottom")
plt.tight_layout()
map_path = "/mnt/data/quartic_concavity_boundary_world_map.png"
plt.savefig(map_path, dpi=250)
plt.show()

spacing_path, map_path
```

For this reason, we also consider transition latitudes below the critical value of \(27.06^\circ\). As a representative example, we adopt the latitude of the Tropic of Cancer (\(23.44^\circ\)) and examine the resulting projection.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq
from scipy.interpolate import PchipInterpolator
from scipy.integrate import cumulative_trapezoid
from matplotlib.collections import LineCollection
import fiona
from shapely.geometry import shape, Polygon, MultiPolygon, LineString, MultiLineString, GeometryCollection

# -----------------------------
# 0. Tropic of Cancer latitude
# Mean obliquity of the ecliptic, IAU/Meeus polynomial, for 2026-05-28.
# epsilon = 23°26'21.448" - 46.8150"T - 0.00059"T^2 + 0.001813*T^3
# T = Julian centuries from J2000.0
# -----------------------------

def julian_day(year, month, day):
    if month <= 2:
        year -= 1
        month += 12
    A = year // 100
    B = 2 - A + A // 4
    return int(365.25*(year + 4716)) + int(30.6001*(month + 1)) + day + B - 1524.5

JD = julian_day(2026, 5, 28)
T_cent = (JD - 2451545.0) / 36525.0
eps_arcsec = 23*3600 + 26*60 + 21.448 - 46.8150*T_cent - 0.00059*T_cent**2 + 0.001813*T_cent**3
tropic_deg = eps_arcsec / 3600

# Use phi1 as Tropic of Cancer
phi1_deg = tropic_deg
phi1 = np.deg2rad(phi1_deg)

# -----------------------------
# 1. Mollweide parallel spacing and derivatives
# -----------------------------

def theta_from_phi(phi):
    phi = np.asarray(phi, dtype=float)
    theta = phi / 2
    for _ in range(70):
        f = 2*theta + np.sin(2*theta) - np.pi*np.sin(phi)
        fp = 2 + 2*np.cos(2*theta)
        theta = np.where(np.abs(fp) > 1e-12, theta - f/fp, theta)
    return theta

def p_moll(phi):
    th = theta_from_phi(phi)
    return np.pi*np.cos(phi)/(2*np.sqrt(2)*np.cos(th))

def derivs_p_moll(phi):
    h = 1e-5
    P = p_moll(phi)
    Q = (p_moll(phi+h) - p_moll(phi-h))/(2*h)
    R = (p_moll(phi+h) - 2*p_moll(phi) + p_moll(phi-h))/(h*h)
    return P, Q, R

# -----------------------------
# 2. Solve phi2 and quartic p(t)=1-a*t^3-b*t^4
#    Matching p, p', p'' to Mollweide at phi2
# -----------------------------

def ab_from(phi1, phi2):
    T = phi2 - phi1
    P, Q, R = derivs_p_moll(phi2)
    M = np.array([[3*T*T, 4*T**3], [6*T, 12*T*T]], dtype=float)
    rhs = np.array([-Q, -R], dtype=float)
    a, b = np.linalg.solve(M, rhs)
    return a, b

def residual_phi2(phi2, phi1):
    T = phi2 - phi1
    if T <= 0:
        return np.nan
    P, Q, R = derivs_p_moll(phi2)
    a, b = ab_from(phi1, phi2)
    return 1 - a*T**3 - b*T**4 - P

def find_phi2(phi1):
    xs = np.deg2rad(np.linspace(np.rad2deg(phi1)+0.2, 88.5, 700))
    vals = np.array([residual_phi2(x, phi1) for x in xs])
    for i in range(len(xs)-1):
        if np.isfinite(vals[i]) and np.isfinite(vals[i+1]) and vals[i]*vals[i+1] < 0:
            return brentq(lambda z: residual_phi2(z, phi1), xs[i], xs[i+1], xtol=1e-13)
    raise RuntimeError("No phi2 root found")

phi2 = find_phi2(phi1)
a, b = ab_from(phi1, phi2)
phi2_deg = np.rad2deg(phi2)

print(f"sinu ≈ {phi1_deg:.10f}°")
print(f"phi1 = {phi1_deg:.10f}°")
print(f"phi2 = {phi2_deg:.10f}°")
print(f"T = {phi2_deg - phi1_deg:.10f}°")
print(f"a = {a:.15g}")
print(f"b = {b:.15g}")
print(f"p(t) = 1 - ({a:.15g}) t^3 - ({b:.15g}) t^4, t in radians")

# -----------------------------
# 3. Plot parallel spacing
# -----------------------------

lat_deg = np.linspace(-90, 90, 6001)
phi = np.deg2rad(lat_deg)
u = np.abs(phi)
t = u - phi1
p_trans = 1 - a*t**3 - b*t**4
p_piece = np.where(u <= phi1, 1.0, np.where(u <= phi2, p_trans, p_moll(phi)))

plt.figure(figsize=(9.5,5.7))
plt.plot(lat_deg, np.ones_like(lat_deg), label="Sinusoidal: 1", linewidth=1.3)
plt.plot(lat_deg, p_moll(phi), label="Mollweide", linewidth=1.3)
plt.plot(lat_deg, p_piece, label="Tropic-boundary quartic transition", linewidth=2.8)
for d in [-phi2_deg, -phi1_deg, phi1_deg, phi2_deg]:
    plt.axvline(d, linestyle="--", linewidth=1)
plt.xlabel("Latitude (degrees)")
plt.ylabel("Parallel spacing function dy/dφ")
plt.title("Parallel spacing with Tropic of Cancer as first transition latitude")
plt.grid(True)
plt.legend()
plt.tight_layout()
spacing_path = "/mnt/data/parallel_spacing_tropic_cancer_quartic_transition.png"
plt.savefig(spacing_path, dpi=220)
plt.show()

# -----------------------------
# 4. Build projection from p(phi): y = ∫p dphi, A = cos(phi)/p
# -----------------------------

def p_spacing(phi):
    phi = np.asarray(phi, dtype=float)
    u = np.abs(phi)
    t = u - phi1
    pt = 1 - a*t**3 - b*t**4
    return np.where(u <= phi1, 1.0, np.where(u <= phi2, pt, p_moll(phi)))

grid_phi = np.linspace(0, np.pi/2 - 1e-7, 30000)
grid_y = cumulative_trapezoid(p_spacing(grid_phi), grid_phi, initial=0)
y_interp = PchipInterpolator(grid_phi, grid_y, extrapolate=True)

def y_of_phi(phi):
    return np.sign(phi)*y_interp(np.abs(phi))

def A_of_phi(phi):
    return np.cos(phi)/p_spacing(phi)

def project_lonlat(lon_deg, lat_deg):
    lon = np.deg2rad(lon_deg)
    ph = np.deg2rad(lat_deg)
    return lon*A_of_phi(ph), y_of_phi(ph)

# -----------------------------
# 5. Geometry helpers and world map
# -----------------------------

def split_ring(coords):
    if len(coords) < 2: 
        return []
    segs, cur = [], [coords[0]]
    prev = coords[0][0]
    for pt in coords[1:]:
        lon = pt[0]
        if abs(lon-prev) > 180:
            if len(cur) >= 2:
                segs.append(cur)
            cur = [pt]
        else:
            cur.append(pt)
        prev = lon
    if len(cur) >= 2:
        segs.append(cur)
    return segs

def densify(coords, max_step=0.75):
    out = []
    for (lon1,lat1),(lon2,lat2) in zip(coords[:-1], coords[1:]):
        out.append((lon1,lat1))
        n = int(max(abs(lon2-lon1), abs(lat2-lat1)) // max_step)
        for k in range(1,n+1):
            f = k/(n+1)
            out.append((lon1+f*(lon2-lon1), lat1+f*(lat2-lat1)))
    out.append(coords[-1])
    return out

def geom_segments(geom):
    segs = []
    if isinstance(geom, Polygon):
        rings = [geom.exterior] + list(geom.interiors)
        for r in rings:
            for part in split_ring(list(r.coords)):
                d = densify(part)
                lon = np.array([p[0] for p in d])
                lat = np.array([p[1] for p in d])
                x,y = project_lonlat(lon, lat)
                segs.append(np.column_stack([x,y]))
    elif isinstance(geom, MultiPolygon):
        for g in geom.geoms:
            segs.extend(geom_segments(g))
    elif isinstance(geom, LineString):
        for part in split_ring(list(geom.coords)):
            d = densify(part)
            lon = np.array([p[0] for p in d])
            lat = np.array([p[1] for p in d])
            x,y = project_lonlat(lon, lat)
            segs.append(np.column_stack([x,y]))
    elif isinstance(geom, MultiLineString):
        for g in geom.geoms:
            segs.extend(geom_segments(g))
    elif isinstance(geom, GeometryCollection):
        for g in geom.geoms:
            segs.extend(geom_segments(g))
    return segs

#world = gpd.read_file(shp_path)
url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
world = gpd.read_file(url)
coast_segments = []
for geom in world.geometry:
    coast_segments.extend(geom_segments(geom))

grat = []
for lon in range(-180, 181, 30):
    lats = np.linspace(-89.999, 89.999, 800)
    lons = np.full_like(lats, lon)
    x,y = project_lonlat(lons,lats)
    grat.append(np.column_stack([x,y]))

for lat in range(-75, 76, 15):
    lons = np.linspace(-180, 180, 1200)
    lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats)
    grat.append(np.column_stack([x,y]))

boundary = []
for lat in [89.999, -89.999]:
    lons = np.linspace(-180,180,1200)
    lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats)
    boundary.append(np.column_stack([x,y]))

for lon in [-180,180]:
    lats = np.linspace(-89.999,89.999,1200)
    lons = np.full_like(lats,lon)
    x,y = project_lonlat(lons,lats)
    boundary.append(np.column_stack([x,y]))

transition_lines = []
for lat in [phi1_deg, -phi1_deg, phi2_deg, -phi2_deg]:
    lons = np.linspace(-180,180,1200)
    lats = np.full_like(lons, lat)
    x,y = project_lonlat(lons,lats)
    transition_lines.append(np.column_stack([x,y]))

fig, ax = plt.subplots(figsize=(13.5,7.2))
ax.add_collection(LineCollection(grat, linewidths=0.45, alpha=0.45))
ax.add_collection(LineCollection(coast_segments, linewidths=0.6))
ax.add_collection(LineCollection(boundary, linewidths=1.0))
ax.add_collection(LineCollection(transition_lines, linewidths=0.95, linestyles="--"))

ax.set_aspect("equal", adjustable="box")
ax.autoscale()
ax.set_title("Equal-area world map with Tropic of Cancer as first transition latitude")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.text(
    0.01, 0.02,
    f"phi1≈{phi1_deg:.5f}° (Tropic), phi2≈{phi2_deg:.5f}°",
    transform=ax.transAxes,
    fontsize=9,
    va="bottom"
)
plt.tight_layout()
map_path = "/mnt/data/tropic_cancer_quartic_transition_world_map.png"
plt.savefig(map_path, dpi=250)
plt.show()

spacing_path, map_path
```

## Conclusion

This study demonstrates that the mid-latitude transition region of the Goode Homolosine projection can be replaced by Strebe-style equal-area polynomial projections while maintaining smooth junctions with the Sinusoidal and Mollweide components.

By expressing the parallel spacing function as a polynomial in latitude and matching it to the Sinusoidal and Mollweide projections, we obtained equal-area projections whose map boundaries satisfy \(C^n\) continuity conditions. Furthermore, admissible parameter ranges were identified in which the boundary retains the desired concavity properties.

These results demonstrate that smooth Homolosine-type projections can be constructed systematically through polynomial control of the parallel-spacing density, yielding projections that are both geometrically well-behaved and visually appealing.
