# A355119: a(n) is the number of n-th order magic triangles composed of the numbers from 1 to n(n+1)/2 which the sum of k-th row and (n-k+1)-th row is same for all k and all three directions, counted up to rotations and reflections.

```python
import itertools
import numpy as np
```

## Firstly, arrange central six numbers to calculate a(5)

```python
i=0
sum3=24
for A in itertools.combinations(range(1,16), 6):
    for B in itertools.permutations(A,6):
        if B[0]>B[2] and B[2]>B[4]:
            if B[0]+B[1]+B[2]==sum3 and B[0]+B[5]+B[4]==sum3 and B[2]+B[3]+B[4]==sum3:
                print(B)
                i+=1
print(i)
```

```python
i=0
Bs=np.zeros([134,6],dtype=int)
sum3=24
for A in itertools.combinations(range(1,16), 6):
    for B in itertools.permutations(A,6):
        if B[0]>B[2] and B[2]>B[4]:
            if B[0]+B[1]+B[2]==sum3 and B[0]+B[5]+B[4]==sum3 and B[2]+B[3]+B[4]==sum3:
                Bs[i,:]=B
                print(B,np.sum(B))
                i+=1
print(i)
```

```python
sum6=sum3*2
i=0
for k in range(134):
    B=Bs[k,:]
    set9=set(range(1,16))-set(B)
    for C in itertools.permutations(set9,6):
        if C[0]+C[1]+C[5]+B[5]+B[3]+C[2]==sum6:
            if C[0]+B[1]+C[3]+B[3]+C[5]+C[4]==sum6:
                if C[1]+B[1]+B[5]+C[4]+C[3]+C[2]==sum6:
                    if C[0]+B[0]+C[5]==C[4]+B[4]+C[3]:
                        if C[4]+B[4]+C[3]==C[1]+B[2]+C[2]:
                            i+=1
print(i)
```
