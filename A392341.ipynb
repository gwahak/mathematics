 # A392341: a(n) is the number of order-n magic triangles composed of the numbers from 0 to (n^2-1) in which the sum of the k-th row and the (n-k)-th row is the same for all k and all three directions, counted up to rotations and reflections.

 # n=1 or 2

trivially, a(1)=1, a(2)=0 

```python
import itertools
```

 # n=3

```python
#        Z2
#        Z1
#      X0   Y0
#      Y1   X1
#   Y2    Z0    X2
```

```python
k=0
for [X0,X1,X2,Y0,Y1,Y2,Z0,Z1,Z2] in itertools.permutations(range(9), 9):
    if X0>Y0:
        if Y0>Z0:
            if X0+Y0+Z1+Z2==18:
                if Y0+Z0+X1+X2==18:
                    if Z0+X0+Y1+Y2==18:
                        k+=1
print(k)
```
